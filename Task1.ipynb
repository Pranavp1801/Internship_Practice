```python
#Question 1

username = "admin"
password = "1234"

inusername = input("Enter username: ")
inpassword = input("Enter password: ")

if inusername == username and inpassword == password:
    print("Login successful")
else:
    print("Login failed")
```

```python
#Question 2

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("\nPass / Fail Analysis:")
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
#Question 3

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("\nCleaned Names List:")
print(cleaned_names)
```

```python
#Question 4

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis:")

for message in messages:
    length = len(message)
    print(f"Message: '{message}' Length: {length}")

    if length > 10:
        print("This message is longer than 10 characters.")
```

```python
#Question 5

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("\nError Log Analysis:")
print("Total ERROR entries:", error_count)
```
