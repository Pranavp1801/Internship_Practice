```python
#Question 1

def check(likes):
  like=0
  for i in likes:
    like+=i
  if like>=1000:
    return "Viral Post"
  else: return "Normal Engagement"

likes=[100,200,300,400]
result=check(likes)
print("Engagement status: ",result)
```

```python
#Question 2
def check(stock):
  if stock<10:
    return "Low Stock alert"
  else: return "Stock sufficient"

n=int(input("Medicine Stock: "))
result=check(n)
print("Status: ",result)
```

```python
#Question 3
def check(rain_data,min_rain):
  total=0
  for i in rain_data:
    total+=i
  if (total/len(rain_data))>=min_rain:
    return "Adequate Rainfall"
  else:
    return "Inadequate Rainfall"

check([70,78,79,80,60,63],75)
```

```python
#Question 4
def check(usernames):
  seen=[]
  duplicate=False
  for name in usernames:
    if name in seen:
      duplicate=True
      break
    else:
      seen.append(name)
  if duplicate:
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Accounts Found: No")

user_data=["userA","userB","userC","userA"]
check(user_data)
```

```python
#Question 5
def check(age):
  print("Patient Age:",age)
  if age>=18:
    print("Eligibility Status: Eligible")
  else:
    print("Eligibility Status: Not Eligible")

age_input=21
check(age_input)
```

```python
#Question 6
def check(prices,limit):
  premium=[]
  for p in prices:
    if p>limit:
      premium.append(p)
  print("Premium Crops:",premium)

prices=[2100,1800,2500,1999,3200]
threshold=2000
check(prices,threshold)
```

```python
#Question 7
def check(errors):
  print("Error Count:",errors)
  if errors==0:
    status="Healthy"
  elif errors<=5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  print("System Status:",status)

errors_detected=7
check(errors_detected)
```

```python
#Question 9
def check(record):
  total=len(record)
  present=0
  for i in record:
    if i==1:
      present+=1
  percent=(present/total)*100
  print("Attendance Percentage:",percent)
  if percent>=75:
    print("Exam Eligibility: Eligible")
  else:
    print("Exam Eligibility: Not Eligible")

record=[1,1,0,1,1]
check(record)
```

```python
#Question 10
def check(units):
  if units<=100:
    total=units*3
  elif units<=200:
    total=(100*3)+((units-100)*5)
  else:
    total=(100*3)+(100*5)+((units-200)*7)

  if total<500:
    category="Low Usage"
  elif total<=1500:
    category="Moderate Usage"
  else:
    category="High Usage"

  return total,category

units_used=250
bill,status=check(units_used)
print("Units Consumed:",units_used)
print("Total Bill:",bill)
print("Usage Category:",status)
```
