```python
#Question 1
s=input("Enter string: ")
words=s.split()
unique_words=set(words)
print("Unique words:", unique_words)
print("Unique words count:", len(unique_words))
```

```python
#Question 2

n=int(input("Enter number of employees: "))
info={}
#Dynamic Input
for i in range(n):
      name=input("Enter name: ")
      salary=int(input("Enter Salary: "))
      info[name]=salary

max_salary=next(iter(info.values())) #set first salary as maximum salary
employee_name=next(iter(info.keys())) #record his name

for i in info:
      if info[i]>max_salary:
            max_salary=info[i]
            employee_name=i

print("Employee with highest salary: ",employee_name)
print(f"Salary of {employee_name} is ",max_salary)
```

```python
#Question 3

n=int(input("Enter number of element: "))
nums=[]
max_num=float('-inf')
min_num=float('inf')
for i in range(n):
      number=int(input("Enter Number: "))
      nums.append(number)

      if number>max_num:
            max_num=number
      if number<min_num:
            min_num=number

print("List: ",nums)
print("Maximum Number: ",max_num)
print("Minimum Number: ",min_num)

```

```python
#Question 4

n=int(input("Enter number of element: "))
count=0
for i in range(n):
  number=int(input(f"Enter Price of element {i+1} :"))
  if number>1000:
    count+=1
print("Products above 1000: ",count)
```

```python
#Question 5

attendance = ["P", "P", "A", "P", "P"]
n=len(attendance)
present=0
for i in attendance:
  if i=="P":
    present+=1

print("Attendance Percentage: ",(present/n)*100)
```

```python
#question 7

s=input("Enter string").lower()
info={}
for i in s:
  if i in info:
    info[i]=info[i]+1
  elif i.isalpha():
    info[i]=1

print(info)
```

```python
#Question 9

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

target=input("Enter Employee search: ").lower()

for i in employees:
    if i.lower()==target:
        print("Employee Found")
        break
else:
    print("Employee Not Found")
```

```python
#Question 10
marks = []

n=int(input("Enter number of subjects: "))

for i in range(n):
    mark=int(input(f"Enter Marks of Subject {i+1}: "))
    marks.append(mark)

average=sum(marks)/len(marks)

print("Average Marks:", average)
```
