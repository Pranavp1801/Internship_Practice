```python
# Question 1
class SmartLight:
    def __init__(self,name):
        self.name=name
        self.status="OFF"
    def turn_on(self):
        self.status="ON"
    def turn_off(self):
        self.status="OFF"
    def display(self):
        print(self.name+" is "+self.status)

light=SmartLight("Bedroom Light")

light.turn_on()
light.display()
```

```python
# Question 2
class Employee:
    def __init__(self,name,id,dept):
        self.name=name
        self.id=id
        self.dept=dept
    def display(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.id)
        print("Department:",self.dept)

e=Employee("Rahul","EMP102","IT")
e.display()
```

```python
# Question 3
class Contact:
    def __init__(self,name,phone):
        self.name=name
        self.phone=phone
    def display(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone:",self.phone)

c=Contact("Anita","9876543210")
c.display()
```

```python
#Question 4
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display(self):
        print("Product:",self.name)
        print("Price: ₹"+str(self.price))

p=Product("Headphones",2499)
p.display()
```

```python
# Question 5
class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def display(self):
        print("Movie:",self.name)
        print("Rating:",self.rating,"/ 5")
m=Movie("Inception",4.8)
m.display()
```
