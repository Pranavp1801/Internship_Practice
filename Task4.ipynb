```python
#Question 1
def mobile_recharge():
  while True:
    value=int(input("Enter the Recharge amount: "))

    if value<50:
      print("Enter Valid Value above 50!")

    elif value==199 or value==299 or value==399 or value==599:
      print("Recharge Successful!")
      break
    else:
      print("Enter Valid Value(199,299,399,599)!")

    choice=input("Do you want to retry(Y/N): ").upper()
    if choice!="Y":
        print("Thank You!")
        break

mobile_recharge()
```

```python
#Question 3
def check_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        print("Average:", average)
        return "Pass"
    else:
        print("Average:", average)
        return "Fail"

student_marks = [60, 70, 45, 80, 50]
result = check_result(student_marks)
print("Result:", result)
```

```python
#Question 4
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak == "Y":
        fare += fare * 0.25   # 25% extra

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (Y/N): ").upper()

    total_fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (Y/N): ").upper()
    if retry != "Y":
        print("Thank you for using Cab Fare Calculator!")
        break
```

```python
#Question 5
def check_attendance(attendance_list):
    present_count = 0
    for status in attendance_list:
        if status == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

records = ["P", "P", "A", "P", "P", "A", "P", "P"]
result = check_attendance(records)
print("Status:", result)
```

```python
#Question 6
def check_password(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

while True:
    pwd = input("Enter password: ")
    result = check_password(pwd)
    print(result)

    choice = input("Do you want to try again? (Y/N): ").upper()
    if choice != "Y":
        print("Thank you!")
        break
```
