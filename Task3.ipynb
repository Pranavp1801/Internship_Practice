```python
#Question 2
s=input("Enter the string: ").lower()
result=[]
word=""
for i in s:
  if i.isalnum():
    word+=i
  if i==" ":
    result.append(word)
    word=""
if word!="":
  result.append(word)

freq={}
for i in result:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
final = {}
for key in freq:
    if freq[key] > 1:
        final[key] = freq[key]

print(final)
```

```python
#Question 4

emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

gmail=0
yahoo=0
n=len(emails)
for i in emails:
  if i[-9:]=="gmail.com":
    gmail+=1
  elif i[-9:]=="yahoo.com":
    yahoo+=1
print("Gmail: ", (gmail/n)*100)
print("Yahoo: ", (yahoo/n)*100)
```
