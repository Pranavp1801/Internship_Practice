```python
import re
import nltk
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python

lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
    if not text or not isinstance(text,str):
        return [],""

    text=re.sub(r'http\S+|www\S+','',text)
    text=re.sub(r'\S+@\S+','',text)
    text=re.sub(r'\d+','',text)

    text=text.lower()

    text=re.sub(r'(.)\1{2,}',r'\1',text)
    text=re.sub(r'[^a-z\s]',' ',text)

    text=re.sub(r'\s+',' ',text).strip()

    tokens=word_tokenize(text)

    clean_tokens=[]
    for word in tokens:
        if len(word)>2 or word in ['no','not']:
            clean_tokens.append(lemmatizer.lemmatize(word))

    clean_sentence=" ".join(clean_tokens)

    return clean_tokens,clean_sentence
```

```python
def full_pipeline(text_list):
    all_tokens=[]
    clean_sentences=[]

    for text in text_list:
        tokens,clean=preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens":all_tokens,
        "clean_sentences":clean_sentences
    }
```

```python
inputs=[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results=[]

for text in inputs:
    tokens,clean=preprocess_text(text)

    total=len(tokens)
    unique=len(set(tokens))
    avg_len=round(sum(len(t) for t in tokens)/total,2) if total>0 else 0

    print("\nOriginal:",text)
    print("Tokens:",tokens)
    print("Clean:",clean)
    print("Total tokens:",total)
    print("Unique tokens:",unique)
    print("Average length:",avg_len)
```

```python
pipeline_output=full_pipeline(inputs)
all_tokens=pipeline_output["tokens"]

freq=Counter(all_tokens)

print("\nTop 10 frequent:",freq.most_common(10))

least=sorted(freq.items(),key=lambda x:x[1])[:5]
print("Top 5 least frequent:",least)
```
