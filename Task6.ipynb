```python
#Question 1
def parking(capacity, logs):
    parked=0
    peak=0

    for log in logs:
        if log=="IN":
            parked += 1
        elif log=="OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full - Alert!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking(capacity, logs)
```

```python
#Question 2
def delivery_time(distance, traffic, weather):
    base_time = distance * 5
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8

    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


delivery_time(8, "High", "Rainy")
```

```python
#Question 3
def theatre_occupancy(total_seats, booked_seats):
    booked_count = 0

    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

total_seats = 200
booked_seats = [1] * 150

theatre_occupancy(total_seats, booked_seats)
```

```python
#Question 4
def server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

```python
#Question 5
def classroom_monitor(resources):
    overused = []
    alert = "No"

    for resource in resources:
        if resources[resource] > 8:
            overused.append(resource)
            alert = "Yes"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_monitor(usage)
```

```python
#Question 6
def event_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

event_registration(100, 105)
```
